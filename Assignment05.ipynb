### 2. 다형성에 대해 설명하고 다형성을 보여주는 자신만의 파이썬 코드 예제를 제시하시오.

```python
class Animal:
    def __init__(self, n):
        self.name = n
    def bark(self):
        pass;

class Dog(Animal):
    def __init__(self, n):
        Animal.__init__(self, n)
    def bark(self):
        print self.name + " 멍멍"

class Cat(Animal):
    def __init__(self, n):
        Animal.__init__(self, n)
    def bark(self):
        print self.name + " 냐옹"
    
d = Dog("멍멍이")
c = Cat("냐옹이")

animalList = []
animalList.append(d);
animalList.append(c);

for animal in animalList:
    animal.bark();
```

### 3. 다음 각 요구사항 모두를 만족시키는 Counter 클래스를 코딩하시오 (정답을 각 요구사항별로 입력할 필요 없이 3번 문제에 대해 1개의 클래스 정의 코드를 제시하면 된다.)
- 요구사항 1. 생성자에 count 값과 step 값을 인자로 받을 수 있다.
- 요구사항 2. 다음과 같이 Counter의 인스턴스를 출력을 해주는 __str__() 메소드를 Counter 클래스 내에 구현하시오.
- 요구사항 3. 다음과 같이 step에 주어진 증분만큼 count를 증가시키는 incr() 메소드를 Counter 클래스 내에 구현하시오.
- 요구사항 4. Counter 클래스 내에 관련 메소드를 추가하여 인스턴스 객체를 직접 호출(call)할 수 있도록 하시오. 인스턴스 객체를 직접 호출했을 때에 내부적으로 incr() 메소드를 호출하는 방법으로 구현하시오.
- 요구사항 5. 다음과 같은 두 개의 산술 연산 (+, -)이 수행될 수 있도록 Counter 클래스 내에 관련 메소드를 추가하시오.
- 요구사항 6. 다음과 같은 관계연산 (>, <, ==)이 수행될 수 있도록 Counter 클래스 내에 관련 메소드를 추가하시오.

```python
class Counter:
    def __init__(self, count, step = 1):
        self.count = count
        self.step = step
    def __str__(self):
        return "[Count (step: %d)] %d" % (self.step, self.count)
    def incr(self):
        self.count += self.step
    def __call__(self):
        self.incr()
    def __add__(self, value):
        return Counter(self.count + value, self.step)
    def __sub__(self, value):
        return Counter(self.count - value, self.step)
    def __lt__(self, other):
        return self.count < other
    def __gt__(self, other):
        return self.count > other
    def __eq__(self, other):
        return self.count == other
    def __ne__(self, other):
        return self.count != other

c = Counter(10)
d = Counter(10, 2)
print c
print d
print

c.incr()
d.incr()
print c
print d
print

c()
d()
print c
print d
print

c = c + 5
d = d - 5
print c
print d
print

print c > 10
print d > 10
print c < 10
print d < 10
print c == 17
print d != 9
```

> 연산자 중복을 사용하여 요구사항을 만족하는 Counter class를 구현하였다.

### 4. 다음은 내장 자료형 list를 서브클래싱하여 만든 MySet 클래스 정의 내용이다. 다음 클래스 정의에서 __init__(), __str()__(), elimicate_duplicate()의 세 개의 메소드 코드 내용을 자신이 다른 사람에게 가르친다고 생각하며 설명해보시오.

```python
class MySet(list):
    def __init__(self, l):
        for e in l:
            self.append(e)
        MySet.eliminate_duplicate(self)
    
    def __str__(self):
        result = "MySet: {"
        for e in self:
            result = result + str(e) + " ,"
        result = result[0:len(result)-2] + "}"
        return result

    @staticmethod    
    def eliminate_duplicate(l):
        s = []
        for e in l:
            if e not in s:
                s.append(e)
        l[:] = []
        for e in s:
            l.append(e)

    
if __name__ == "__main__":
    s = MySet([1, 2, 2, 3])
    print s
    t = MySet([2, 3, 4, 5, 6, 7, 8, 8, 8, 8, 8, 9])
    print t
```

> - \_\_init\_\_은 생성자로, MySet Class를 생성할 때 자동으로 호출된다. 3~4라인을 보면 l 이라는 파라미터를 받아 self에 넣는 것을 알 수 있는데 이 MySet이라는 Class는 list를 상속받았기 때문에 그 자체가 리스트이다. 따라서 self.append(e)와 같은 형태가 가능하다. 5라인에서는 slininate_duplicate라는 함수를 호출하고있다.
- \_\_str\_\_은 print 예약어나 str() 내장함수 호출에 대응되는 메소드이다. 8라인에서 중괄호 앞부분을 작성하고 9~10에서 보유한 요소들을 반복해서 " ,"와 함께 찍어낸다. 11라인에서는 0부터 뒤에서-2번째까지를 잘라내고 중괄호를 닫는것을 볼 수 있는데, 뒤에서 2번째를 자르는 이유는 for in문을 통해 마지막 요소 뒤에서 공백쉼표가 찍혔기 때문이다. 이를 지우고 완성된 문자열을 반환한다.
- eliminate_duplicate는 입력받은 리스트의 중복을 제거하는 함수이다. 그런데 그냥 중복을 제거한 리스트를 반환하는 것이 아니라 중복을 제거한 리스트(s)를 다시 l에 집어넣는 것을 알 수 있다. 따라서 이 함수를 호출하면 a = eliminate_duplicate(a)가 아닌 eliminate_duplicate(a) 만으로도 a는 중복이 제거된 리스트가 될 것이다. 또한 이 메소드는 정적 메소드로 선언되었기 때문에 꼭 인스턴스 객체가 아니더라도 클래스 이름만으로도 호출하여 사용이 가능하다.

### 5. 4번 문제에 정의된 MySet 클래스에 메소드를 추가하여 다음 각 요구사항 모두를 만족시키는 코딩을 제시하시오
- 요구사항 1. | 연산으로 두 집합의 합집합을 반환한다.
- 요구사항 2. & 연산으로 두 집합의 교집합을 반환한다.
- 요구사항 3. - 연산으로 두 집합의 차집합을 반환한다.

```python
class MySet(list):
    def __init__(self, l):
        for e in l:
            self.append(e)
        MySet.eliminate_duplicate(self)
    
    def __str__(self):
        result = "MySet: {"
        for e in self:
            result = result + str(e) + " ,"
        result = result[0:len(result)-2] + "}"
        return result

    @staticmethod    
    def eliminate_duplicate(l):
        s = []
        for e in l:
            if e not in s:
                s.append(e)
        l[:] = []
        for e in s:
            l.append(e)
        
    # or
    def __or__(self, l):
        s = []
        for e in self:
            s.append(e)
        for e in l:
            s.append(e) # self와 l을 합친 다음
        MySet.eliminate_duplicate(s) # 중복을 제거한다.
        return s
            
    # and
    def __and__(self, l):
        s = []
        for e in self:
            if e in l:
                s.append(e) # self와 l에 공통으로 있는 맴버만 추가
        return s
    
    # sub
    def __sub__(self, l):
        s = []
        for e in self:
            if e not in l:
                s.append(e) # self에는 있고, l에는 없는 맴버만 추가
        return s
        
if __name__ == "__main__":
    s = MySet([1, 2, 2, 3])
    print s
    t = MySet([2, 3, 4, 5, 6, 7, 8, 8, 8, 8, 8, 9])
    print t

u = s | t
print u

u = s & t
print u

s = MySet([1, 2, 3]) 
t = MySet([3, 4, 5]) 
u = s - t 
print u 
```

> MySet이 마치 set과 같이 동작하도록 연산자 중복을 사용하여 구현하였다. 코드 동작이 간단하여 주석으로 설명하였다.

### 8. (오일러 17) 1부터 5까지의 숫자를 영어로 쓰면 one, two, three, four, five 이고, 각 단어의 길이를 더하면 3 + 3 + 5 + 4 + 4 = 19 이므로 사용된 글자는 모두 19개입니다. 1부터 1,000까지 영어로 썼을 때는 모두 몇 개의 글자를 사용해야 할까요?

```python
# 각 숫자를 알파벳으로 썻을 때의 문자 길이
numLen={0:0,1:3,2:3,3:5,4:4,5:4,6:3,7:5,8:5,9:4,10:3,11:6,12:6,13:8,14:8,15:7,16:7,17:9,18:8,19:8,20:6,30:6,40:5,50:5,60:5,70:7,80:6,90:6,}

sumv = 0
for i in range(1000):
    under = i%100
    hundred = i/100
    
    if under < 20:
        sumv += numLen[under]
    else:
        digit1 = under % 10
        digit10 = under - digit1
        sumv += (numLen[digit1] + numLen[digit10])
    
    if hundred < 20:
        sumv += numLen[hundred]
    else:
        digit1 = hundred % 10
        digit10 = hundred - digit1
        sumv += (numLen[digit1] + numLen[digit10])
    
    if hundred > 0:
        sumv += 7 # "hundred" 글자
        
    if under != 0 and hundred != 0:
        sumv += 3 # "and" 추가

sumv += 11 # "one thousand"
print sumv
    
```

> 우선 각 숫자를 영어로 썻을 때의 각 자리수를 미리 구하여 저장하였다. 그리고 100의 자리와 그 이하의 자리수를 따로 구분하여 미리 저장된 글자수를 불러와 더하였고, 만약 100의 자리가 존재할 경우 7글자(hundred)를 더하고, 100의 자리와 그 이하의 자리가 모두 존재하는 경우에는 and로 묶어주어야 하므로 3글자(and)를 추가로 더하였다. 마지막으로 1000은 추가로 더하였다.

```python
import urllib2
import string

stopword = "a,about,above,across,after,again,against,all,almost,alone,along,already,also,although,always,among,an,and,another,any,anybody,anyone,anything,anywhere,are,area,areas,around,as,ask,asked,asking,asks,at,away,b,back,backed,backing,backs,be,became,because,become,becomes,been,before,began,behind,being,beings,best,better,between,big,both,but,by,c,came,can,cannot,case,cases,certain,certainly,clear,clearly,come,could,d,did,differ,different,differently,do,does,done,down,down,downed,downing,downs,during,e,each,early,either,end,ended,ending,ends,enough,even,evenly,ever,every,everybody,everyone,everything,everywhere,f,face,faces,fact,facts,far,felt,few,find,finds,first,for,four,from,full,fully,further,furthered,furthering,furthers,g,gave,general,generally,get,gets,give,given,gives,go,going,good,goods,got,great,greater,greatest,group,grouped,grouping,groups,h,had,has,have,having,he,her,here,herself,high,high,high,higher,highest,him,himself,his,how,however,i,if,important,in,interest,interested,interesting,interests,into,is,it,its,itself,j,just,k,keep,keeps,kind,knew,know,known,knows,l,large,largely,last,later,latest,least,less,let,lets,like,likely,long,longer,longest,m,made,make,making,man,many,may,me,member,members,men,might,more,most,mostly,mr,mrs,much,must,my,myself,n,necessary,need,needed,needing,needs,never,new,new,newer,newest,next,no,nobody,non,noone,not,nothing,now,nowhere,number,numbers,o,of,off,often,old,older,oldest,on,once,one,only,open,opened,opening,opens,or,order,ordered,ordering,orders,other,others,our,out,over,p,part,parted,parting,parts,per,perhaps,place,places,point,pointed,pointing,points,possible,present,presented,presenting,presents,problem,problems,put,puts,q,quite,r,rather,really,right,right,room,rooms,s,said,same,saw,say,says,second,seconds,see,seem,seemed,seeming,seems,sees,several,shall,she,should,show,showed,showing,shows,side,sides,since,small,smaller,smallest,so,some,somebody,someone,something,somewhere,state,states,still,still,such,sure,t,take,taken,than,that,the,their,them,then,there,therefore,these,they,thing,things,think,thinks,this,those,though,thought,thoughts,three,through,thus,to,today,together,too,took,toward,turn,turned,turning,turns,two,u,under,until,up,upon,us,use,used,uses,v,very,w,want,wanted,wanting,wants,was,way,ways,we,well,wells,went,were,what,when,where,whether,which,while,who,whole,whose,why,will,with,within,without,work,worked,working,works,would,x,y,year,years,yet,you,young,younger,youngest,your,yours".split(',')

class WebWordsFrequency:
    def __init__(self, *l):
        self.urlList = []
        self.words = {}
        for url in l:
            self.addUrl(url)
            
    def addUrl(self, url):
        self.getWords(url)
        self.urlList.append(url)
        
    def removeUrl(self, url):
        if url in self.urlList:
            del self.words[url]
            self.urlList.remove(url)
            
    def listUrls(self):
        for url in self.urlList:
            print url
            
    # url을 조사하여 사전 리스트를 만드는 함수
    def getWords(self, url):
        self.words[url] = {}
        source = urllib2.urlopen(url).read()
        text = ''

        # 주석 제거
        while '<!--' in source:
            # 주석은 <!-- 과 --> 의 쌍으로 나타내어지므로 index 함수를 사용해 주석을 찾아 없앤다.
            left = source.index('<!--')
            right = source.index('-->')
            source = source[:left] + source[right+3:]

        # 태그가 없어질때까지 반복
        while '<' in source:
            # 태그는 < 과 > 의 쌍으로 나타내어지므로 index 함수를 사용해 태그 부분을 찾아 없앤다.
            left = source.index('<')
            right = source.index('>')
            text = text+' '+source[:left]
            source = source[right+1:]

        # 불용어를 제외한 단어 저장
        for t in text.split():
            for char in string.punctuation:
                t = t.replace(char, '')
            t = t.lower() # 대소문자를 구분하지 않고 단어를 저장한다.
            if len(t) > 0 and not t in stopword:
                if not t in self.words[url].keys():
                    self.words[url][t]=1
                else:
                    self.words[url][t]=self.words[url][t]+1
                        
    def getWordsFrequency(self):
        allWords = {}
        # 각각 따로 저장된 각 사이트 별 사전을 합쳐서 반환
        for url, wordDic in self.words.items():
            for word in wordDic.keys():
                if not word in allWords.keys():
                    allWords[word] = wordDic[word]
                else:
                    allWords[word] += wordDic[word]
                
        return allWords
    
    def getMaxFreqencyWords(self):
        if len(self.urlList) == 0:
            return None
        
        allWords = self.getWordsFrequency() # 합쳐진 사전을 구함
        l = []
        # 가장 많이 등장한 단어를 구하고
        maxv = 0
        for word, count in allWords.items():
            if count > maxv:
                maxv = count;
        # 리스트에 저장
        for word, count in allWords.items():
            if count == maxv:
                l.append(word)
        return l
    
    def searchUrlByWord(self, keyword):
        similarity = {}
        l = []
        
        for url in self.urlList:
            # 유사도는 기본 10점
            similarity[url] = 10
            if keyword in self.words[url].keys():
                # 연관 단어 수만큼 유사도 보너스
                similarity[url] += self.words[url][keyword]
            # 이곳에 더욱 추가하자면, 키워드와 연관된 복수개의 단어를 구해 추가적인 보너스 점수를 준다면 더욱 정확한 결과를 얻을 수 있을 것이다.
        # 가장 유사한 사이트를 구하고
        maxv = 0
        for url, count in similarity.items():
            if count > maxv:
                maxv = count;
        # 리스트에 저장
        for url, count in similarity.items():
            if count == maxv:
                l.append(url)
        return l
        
```

### 11. 위 10번 문제에서 정의한 WebWordsFrequency 클래스를 상속하여 OrderedWebWordsFrequency 클래스를 정의하고 슈퍼클래스에 정의된 getWordsFrequency() 메소드를 오버라이드 하여 단어 출현 빈도를 내림 차순으로 정렬하여 리스트로 출력하시오.

```python
class OrderedWebWordsFrequency(WebWordsFrequency):
    def __init__(self, *l):
        WebWordsFrequency.__init__(self, *l)
    
    def getWordsFrequency(self, reverse = False):
        allWords = WebWordsFrequency.getWordsFrequency(self)
        
        # 내림차순을 기본으로 정렬 후 반환
        l = []
        tmp = allWords.values()
        tmp.sort(reverse = not reverse)
        for value in tmp:
            for key in allWords.keys():
                if allWords[key] == value:
                    l.append((key, allWords[key]))
                    del allWords[key]
                    break
        return l
    def __iter__(self):
        self.i = -1
        self.tmp = self.getWordsFrequency()
        return self
    def next(self):
        self.i+=1
        if len(self.tmp) <= self.i:
            raise StopIteration
        return self.tmp[self.i]
    #def __getitem__(self, n):
    #   l = self.getWordsFrequency()
    #  if len(l) <= n:
    #     raise StopIteration
    #return l[n]
```

> 메서드를 오버라이드하여 (키, 값)쌍으로 정렬된 리스트를 반환하도록 하였다. allWords(모든 url 단어 사전을 합친 사전)을 구하는 부분은 부모의 함수를 그대로 사용하였고, 정렬하여 리스트를 반환하는 부분만 추가하여 코드를 적극적으로 재사용하였다.
